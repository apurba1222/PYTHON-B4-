```python
import pandas as pd

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/seattle-weather.csv")
print(data.head())
```

```python

df = pd.DataFrame(data)
df
# Count occurrences of each weather type
weather_counts = df["weather"].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(weather_counts, labels=weather_counts.index, autopct='%1.1f%%', colors=["skyblue", "orange", "green", "red", "purple"])
plt.title("Weather Distribution in Seattle")
plt.show()
```
